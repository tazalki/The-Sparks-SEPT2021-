# Checking The Profits In Different States

```python
print("total number of states are:",data['State'].nunique)
print(data['State'].value_counts())
```

```python
States_profit = data.groupby('State').sum()['Profit'].sort_values(ascending=False).reset_index()
#sorting the states by their profits in descending order(from highest to lowest profit.)

plt.figure(figsize=(15,10))
plt.title('Most Profitable States')
sns.barplot(x = 'State',y = 'Profit', data = States_profit.head(15)) #printing first 15 highest profitable states
plt.xticks(rotation = 60 , fontsize = 15) #rotating the labels on x-axis 
plt.yticks(fontsize = 15) 
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.title('Least Profitable States')
sns.barplot(x = 'State',y = 'Profit', data = States_profit.tail(15)) #printing 15 lowest profitable states
plt.xticks(rotation = 60 , fontsize = 15)
plt.yticks(fontsize = 15)
plt.show()
```

Here we can predict that, State with highest profit is California and State with lowest profit is Texas

# Checking Profits in different Cities

```python
print('Total no. of Cities are : ',  data['City'].nunique())
print(data['City'].value_counts())
```

```python
city_profit = data.groupby('City').sum()['Profit'].sort_values(ascending = False).reset_index()
#grouping the cities to check the profits in each city and sorting them in descending order from highest to lowest profit.

plt.figure(figsize=(15,10))
plt.title('Most Profitable Cities')
# sns.barplot(x = city_profit['City'][:15],y = city_profit['Profit'][:15])
sns.barplot(x = 'City',y = 'Profit',data = city_profit.head(15))
plt.xticks(rotation = 60 , fontsize = 15)
plt.yticks(fontsize = 15)
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.title('Least Profitable Cities')
sns.barplot(x = 'City', y = 'Profit', data = city_profit.tail(15)) 
plt.xticks(rotation = 60 , fontsize = 15)
plt.yticks(fontsize = 15)
plt.show()
```

Here we can predict that, City with highest profit is New York and city with lowest profit is Philadelphia

# Checking Sales in different states

```python
State_sales = data.groupby('State').sum()['Sales'].sort_values(ascending = False).reset_index()

plt.figure(figsize=(15,10))
plt.title('States with highest sales rate')
sns.barplot(x = 'State', y = 'Sales', data = State_sales.head(15)) 
plt.xticks(rotation = 60 , fontsize = 15)
plt.yticks(fontsize = 15)
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.title('States with lowest sales rate')
sns.barplot(x = 'State', y = 'Sales', data = State_sales.tail(15)) 
plt.xticks(rotation = 60 , fontsize = 15)
plt.yticks(fontsize = 15)
plt.show()
```

So from this we can predict that the state having highest sales rate is California and State having lowest sales rate is North Dakota

# Sales vs Profit

```python
Sales = data.groupby('State').sum()['Sales'].nlargest(n=15) #top 15 states with highest sales
Profits = data.groupby('State').sum()['Profit'].nlargest(n=15) #top 15 states with highest sales

Sales.plot(kind='bar', figsize=(10,5),fontsize = 15, color = 'lightgreen')
Profits.plot(kind='bar', figsize=(10,5),fontsize = 15 , color='Purple')
plt.xlabel('States',fontsize = 15)
plt.ylabel('Total sales and profit',fontsize = 15)
plt.title("Sales vs Profit",fontsize = 15)
plt.show()
```

From above plot, we can conclude that California and New York has highest sales and Profits
